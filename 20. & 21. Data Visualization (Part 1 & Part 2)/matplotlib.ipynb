```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $m

# Plot
plt.plot(years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years, oscar_revenue, color="red", marker="<", linestyle="-.", linewidth="2", label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

plt.legend()
```

```python
with plt.xkcd():
    plt.plot(years, oscar_revenue, label="Oscar Movies")
    plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")
    
    plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
    plt.xlabel("Year")
    plt.ylabel("Revenue (in $M)")
    
    plt.legend()
    plt.show()
```

```python
# Bar charts

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $m

plt.bar(years, oscar_revenue, color="Green")
plt.title("Revenue in each Year for Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

for i in range(len(years)):         
    plt.text(years[i], oscar_revenue[i]+20, str(oscar_revenue[i]), ha="center")

plt.ylim(0, max(oscar_revenue)+100)
```

```python
import numpy as np

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232] # in $m

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar Movies") # Shifting to LEFT
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar Movies")# Shifting to RIGHT

plt.title("Oscar vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
plt.legend()

plt.xticks(x, years)
```

```python
# Scatter Plots

people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# colors = ["red", "green", "blue", "orange", "purple", "red", "green", "blue", "orange", "purple"]
colors = ["green" if x < 135 else "red" for x in bp]

plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i]+1, bp[i]+1))

plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

# colors = ["red", "black", "pink", "blue", "yellow"]

explode = [0, 0, 0, 0.2, 0]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
       wedgeprops = {
            "edgecolor": "black",
            "linewidth": 1,
            "linestyle": "--",
            # "fill": False
       },
        explode=explode,
        startangle=360,
        shadow=True
       )
plt.show()
```

```python
# Histogram

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

scores = np.random.normal(70, 10, 100) # Normal Distribution
bins = [0, 30, 70, 90, 100]

plt.hist(scores, color="skyblue", edgecolor="black", bins=bins)
plt.ylabel("# of students")
plt.xlabel("scores")
print(scores)
```

```python
# Multiple datasets
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

# Transaction Amount Distribution - Legit vs Fraud
plt.hist(fraud_transactions, color="red", bins=20, label="Fraud", edgecolor="black", alpha=0.5)
plt.hist(legit_transactions, color="green", bins=20, label="Legit", edgecolor="black", alpha=0.5)

plt.title("Transaction Amount Distribution - Legit vs Fraud")
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
```
